```python
age=pd.get_dummies(df['Age'],drop_first=True)
fd=pd.concat([df,age],axis=1)
```

```python
X=df[['EstimatedSalary','Age']]
y=df['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(X)
```

```python
sf=ss.transform(X)
sf
```

```python
df_feat=pd.DataFrame(sf,columns=X.columns)
df_feat
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(sf,y,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
pred=knn.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

This is all about K-Nearest Neighbour and it's classifier.

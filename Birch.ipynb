```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.datasets import make_blobs
from sklearn.cluster import Birch
```

```python
X,clusters=make_blobs(n_samples=150,centers=6,cluster_std=1.6,random_state=0)
plt.scatter(X[:,0],X[:,1],alpha=0.5,edgecolors='b')
```

```python
brc=Birch(branching_factor=60,n_clusters=None,threshold=1.5)
```

```python
brc.fit(X)
```

```python
labels=brc.predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=labels,cmap='rainbow',alpha=0.6,edgecolors='b')
```

```python
plt.scatter(X[:,0],X[:,1],c=labels,cmap='rainbow',alpha=0.5,edgecolors='b')
```

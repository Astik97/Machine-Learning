```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
```

```python
print(cancer['DESCR'])
```

```python
data=cancer
```

```python
data
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
X_train,X_test=train_test_split(df,test_size=0.3,random_state=0)
sc=StandardScaler()
sc.fit(X_train)
X_train_std=sc.transform(X_train)
X_test_std=sc.transform(X_test)
```

```python
pca=PCA()
X_train_pca=pca.fit_transform(X_train_std)
exp_var_pca=pca.explained_variance_ratio_
cum_sum_eigenvalues=np.cumsum(exp_var_pca)
plt.bar(range(0,len(exp_var_pca)),exp_var_pca,alpha=0.5,align='center',label='Individual Exp Variance')
plt.step(range(0,len(cum_sum_eigenvalues)),cum_sum_eigenvalues,where='mid',label='Cumulative Exp Variance')
plt.ylabel('Explained Variance Ratio')
plt.xlabel('PC Index')
plt.legend()
plt.tight_layout()
plt.show()
```

Experience Variance Ratio by PCA,Based on Variance Ratio,we can select appropriate Principal Component Value

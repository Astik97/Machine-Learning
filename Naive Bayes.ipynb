```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
from sklearn.datasets import make_blobs
X,y=make_blobs(100,5,centers=6,random_state=5,cluster_std=1.6)
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap='RdBu')
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X,y)
```

```python
rng = np.random.RandomState(0)
Xnew = np.array([6, 10, 0, 0, 0] + [10, 15, 0, 0, 0] * rng.rand(6000, 5))
ynew = model.predict(Xnew)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap='RdBu')
lim=plt.axis()
plt.scatter(Xnew[:,0],Xnew[:,1],c=ynew,s=29,cmap='RdBu',alpha=0.1)
plt.axis(lim);
```

```python
yprob=model.predict_proba(Xnew)
yprob[8:].round(3)
```

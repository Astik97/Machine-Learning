```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
```

```python
data=pd.read_csv("A:\Dataset\pokemon.csv\pokemon.csv")
data
```

```python
data.drop(['Type 1','Type 2','Legendary'],axis=1,inplace=True)
data
```

```python
plt.scatter(data['Attack'],data['Defense'])
```

Scatter plot of Attack and Defense.

```python
plt.scatter(data['Speed'],data['Generation'])
```

Scatter plot of Speed and Generation.

```python
X=data.iloc[:,:1].values
y=data.iloc[:,1].values
```

```python
nbrs=NearestNeighbors(n_neighbors=5)
nbrs.fit(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=6)
```

```python
dist,index=nbrs.kneighbors(X)
plt.plot(dist.mean(axis=1))
```

Mean Distance from the K-Neighbours.

```python
outlier_index=np.where(dist.mean(axis=1)>0.10)
outlier_index
```

```python
outlier_values=data.iloc[outlier_index]
outlier_values
```

```python
plt.scatter(data['Attack'],data['Defense'],color='black',s=65)
plt.scatter(outlier_values['Attack'],outlier_values['Defense'],color='blue')
```

K-Nearest Neighbours with Anomalies.
